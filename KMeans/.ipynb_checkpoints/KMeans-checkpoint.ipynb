```python
X= data.iloc[:,[3,4]].values
```

```python
m=X.shape[0]
n=X.shape[1]
n_iter=0
K=5 # 5 cluster centres
```

```python
centroids=np.array([]).reshape(n,0)
for i in range(K):
    rand=rd.randint(0,m-1)
    #print(X[rand])
    centroids=np.c_[centroids,X[rand]]
    #print(centroids)
```

```python
output={}
euclidean_distance=np.array([]).reshape(m,0)
for k in range(K):
    print(centroids[:,k])
    sqrt=X-centroids[:,k]
    
    dist=np.sum((X-centroids[:,k]**2),axis=1)
    #print(dist.shape)
    euclidean_distance=np.c_[euclidean_distance,dist]
C=np.argmin(euclidean_distance,axis=1)
```
